```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
print("We prpeare a dataframe of the given data of print-head in millions")
```

```python
print("We calculate the sample statisctical parameters")
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)
print(f"Sample Mean: {mean:.2f}")
print(f"Sample Standard Deviation: {std_dev:.2f}")
print(" ")
print("Since n<30, we will use student's T test for calculating the confidence levels")
print("For 99% confidence interval(CI) alpha would be 0.01./n Additionally, using degree of freedom(df) we calculate the critical value for t test")
alpha = 0.01
df = n - 1
t_critical = stats.t.ppf(1 - alpha / 2, df)

print("Next we calculate the margin of error")
margin_of_error = t_critical * (std_dev / np.sqrt(n))
print(f"margin of error: {margin_of_error:.2f}")

print(" ")
print("Using margin of error, we calculate the upper and lower CI intervals")
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error
print(f"99% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")


sns.distplot(data, hist=True)
plt.axvline(ci_lower, color='red', linestyle='--', label=f'Lower bound {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'Upper bound {ci_upper:.2f}')
plt.axvline(mean, color='blue', label=f'Mean {mean:.2f}')
plt.title('Print-head Durability Distribution with 99% Confidence Interval')
plt.legend()
plt.xlabel('Durability (millions of characters)')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Given population standard deviation= 0.2")
pop_std_dev = 0.2

print("Alpha remains same, we calculate the z critical value")
z_critical = stats.norm.ppf(1 - alpha / 2)
print(f"Z critical value: {z_critical:.2f}")

print(" ")
print("Similarly we calculate the margin of error")
moe_pop = z_critical * (pop_std_dev / np.sqrt(n))
print(f"Margin of error: {moe_pop:.2f}")
print(" ")
print("We calculate the CI for population distribution")
ci_lower_pop = mean - moe_pop
ci_upper_pop = mean + moe_pop
print(f"99% Confidence Interval (Known Population Std Dev): ({ci_lower_pop:.2f}, {ci_upper_pop:.2f})")
```

```python
sns.distplot(data, hist=True)
plt.axvline(ci_lower_pop, color='red', linestyle='--', label=f'Lower bound {ci_lower_pop:.2f}')
plt.axvline(ci_upper_pop, color='green', linestyle='--', label=f'Upper bound {ci_upper_pop:.2f}')
plt.axvline(mean, color='blue', label=f'Mean {mean:.2f}')
plt.title('Print-head Durability Distribution with 99% Confidence Interval (Population Std Dev Known)')
plt.legend()
plt.xlabel('Durability (millions of characters)')
plt.ylabel('Frequency')
plt.show()
```
